# IT Security Agent

## 1. Introduction

## 1.1 Background

Software vulnerabilities pose significant security risks, especially in modern software ecosystems that depend on numerous third-party components. This project develops an AI-based IT Security Agent that identifies known vulnerabilities from software images or Software Bill of Materials (SBOMs) using the National Vulnerability Database (NVD). In addition to implementing the detection workflow, the project follows Responsible AI principles by analyzing the dataset, evaluating potential risks and limitations, discussing relevant regulations, and providing a pseudo model card to assess the system from both technical and ethical perspectives.

## 1.2 Regulatory Analysis

To ensure compliance with European regulations and Responsible AI principles, our system is designed within strict legal boundaries:

**EU AI Act Classification (Minimal/Limited Risk):** The system does not autonomously control or patch critical digital infrastructure, exempting it from Annex III High-Risk categories. It acts purely as a decision-support tool designed to improve the results of a human-led activity (vulnerability discovery).

**GDPR & Data Privacy:** The agent processes only technical documentation (SBOMs/Software Images) and open-source vulnerability feeds (CVEs, CPEs, CVSS scores). It explicitly does not collect or infer any Personally Identifiable Information (PII), operating entirely outside GDPR constraints.

**Ethical Accountability:** To address the ethical challenge of AI responsibility, the system enforces a strict Human-in-the-Loop architecture. The AI flags potential threats, but the final security decisions and remediation remain the absolute responsibility of the human user.

## 2. Dataset

### 2.1 NVD Dataset

The National Vulnerability Database (NVD) is the U.S. government's official repository of publicly disclosed software vulnerabilities. Each vulnerability is identified by a unique CVE identifier and includes detailed information such as descriptions, affected products, severity scores (CVSS), weakness classifications (CWE), affected configurations (CPE), and external references. The NVD data feeds are updated regularly and provide a standardized foundation for vulnerability analysis and security research.

### 2.2 Dataset Overview

The NVD data is distributed in JSON format, where each record represents a single CVE entry containing structured vulnerability information. For this project, relevant fields including the CVE identifier, publication date, affected vendors and products, CPEs, CWEs, CVSS metrics, SSVC information, and references are extracted and stored in MongoDB to support efficient querying and vulnerability matching.

The complete dataset consists of 362,810 CVE records spanning multiple years (2002-2026). This large-scale dataset provides comprehensive coverage of publicly disclosed vulnerabilities and serves as the foundation for the exploratory data analysis and the development of the IT Security Agent presented in this project.

### 2.3 Dataset Limitations & Biases

**Reporting Lag:** There is a known time delay between the discovery of a vulnerability and its official publication on the NVD. Consequently, the agent is inherently blind to zero-day vulnerabilities.

**Severity Bias:** NVD relies on CVSS metrics which represent a theoretical, base severity. These scores do not account for the specific context or actual exploitability within a user's unique environment, which can lead to risk overestimation.

**CPE Mismatching:** Accurately mapping software names extracted from SBOMs to NVD's strict Common Platform Enumeration (CPE) format is challenging. Inconsistencies in naming conventions can result in False Positives or missed detections.

## 4. Project Plan

The project is divided into four main stages.
- First, the NVD data is collected, parsed, and stored in a structured MongoDB database for efficient querying.
- Second, exploratory data analysis (EDA) is performed to understand the characteristics of the dataset and identify relevant information for vulnerability matching.
- Third, an AI-based IT Security Agent is developed to extract software information from software images or SBOMs and match it against the NVD database to identify known vulnerabilities.
- Finally, the system is evaluated from a Responsible AI perspective, including regulatory analysis, risk assessment, explainability, limitations, and a pseudo model card.

## 5. System Architecture

The proposed system consists of four main components: data management, input processing, vulnerability matching, and AI-assisted reporting.

First, the NVD data feeds are parsed and stored in a MongoDB database, where relevant fields such as CVE identifiers, affected products, CPEs, CWEs, CVSS metrics, and references are indexed for efficient retrieval.

Next, the agent accepts either a Software Bill of Materials (SBOM) or a software image as input. For software images, OCR and image analysis are used to extract software names and versions. For SBOMs, package information is parsed directly.

The extracted software information is then matched against the vulnerability database using package names, versions, and CPE information to identify related CVEs. Finally, a Large Language Model (LLM) summarizes the matched vulnerabilities, explains their potential security impact, and generates a structured security report for the user.

## 7. Risk Assessment

| Risk                               | Description                                                                        | Mitigation                                                                       |
|------------------------------------|------------------------------------------------------------------------------------|----------------------------------------------------------------------------------|
| Incorrect Software Identification  | OCR or SBOM parsing may extract incorrect package names or versions.               | Validate extracted packages and allow manual review.                             |
| Incomplete Vulnerability Detection | Missing identifiers or unsupported version ranges may prevent matching known CVEs. | Use multiple identifiers (CPE, PURL, package name) and improve version matching. |
| Outdated Vulnerability Database    | Newly disclosed vulnerabilities may not yet exist in the local database.           | Regularly synchronize the NVD database.                                          |
| AI-generated Misinterpretation     | The LLM may oversimplify or misinterpret vulnerability information.                | Limit the LLM to summarization and require human review.                         |
| Adversarial Input                  | Malicious or manipulated images/SBOMs may intentionally hide vulnerable software.  | Validate input format and perform consistency checks before reporting.           |

## 8. Fairness Analysis

### Fairness Considerations

- Input Modality Consistency
    - The system should provide similar results for SBOM files and software screenshots whenever they describe the same software.
- Software Ecosystem Coverage
    - Vulnerability coverage may differ across software ecosystems because some ecosystems have more complete security data than others.
- Identifier Availability
    - Software with standardized identifiers (such as CPE or PURL) is easier to match than software with incomplete metadata.
- Vendor Reporting Differences
    - Some vendors disclose vulnerabilities more actively than others, resulting in uneven vulnerability coverage.
- Consistent AI Reporting
    - The LLM should generate reports using the same evaluation criteria for all software, regardless of vendor, ecosystem, or input source.

**This project does not make decisions about people. Fairness therefore refers to providing consistent vulnerability analysis across different software, vendors, and input formats.**

## 11. Conclusion

We developed an AI-assisted vulnerability analysis system that automatically analyzes software dependencies from SBOM files or screenshots of dependency lists. The system combines OCR, vulnerability matching, version filtering, and LLM-based report generation to produce readable security reports. We evaluated the system from multiple perspectives, including performance, reliability, fairness, and explainability. The results show that the proposed pipeline is transparent, practical, and suitable for supporting software security analysis. Although the current implementation has limitations, such as relying on the NVD database and supporting a limited range of input formats, it provides a solid foundation for future extensions, including additional vulnerability databases, improved OCR capabilities, and broader software ecosystem support. Overall, this project demonstrates that AI can effectively reduce manual effort in vulnerability assessment while keeping the decision-making process transparent and traceable.
