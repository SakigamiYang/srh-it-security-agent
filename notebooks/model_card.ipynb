# Model Card

## Model Details

| Item             | Description                                                   |
|------------------|---------------------------------------------------------------|
| Model Name       | IT Security Agent                                             |
| Version          | 1.0                                                           |
| Model Type       | Retrieval-Augmented AI Agent                                  |
| Main Components  | RapidOCR, Vulnerability Matching Engine, Large Language Model |
| Knowledge Source | NVD CVE Database                                              |
| Output           | Markdown Security Report                                      |

## Intended Use

The system is designed to assist software developers and security analysts by automatically identifying known vulnerabilities in software dependencies and generating concise security reports.

Supported inputs:

- SPDX SBOM (JSON)
- Screenshot of dependency lists (e.g., requirements.txt)

The system is intended to support vulnerability analysis and should not replace professional security assessment.

## Model Pipeline

1. Parse the input (SBOM or screenshot).
2. Extract package names and versions.
3. Search the local vulnerability database.
4. Filter vulnerabilities using version constraints.
5. Generate a Markdown report with an LLM.

## Training Data

The system does not train a predictive model for vulnerability detection.

It relies on:

- National Vulnerability Database (NVD)
- CVE metadata
- CVSS metrics
- SSVC information (when available)

The auxiliary explainability model is trained using extracted CVE metadata solely for SHAP analysis.

## Performance

The system was evaluated using several software dependency examples.

Evaluation includes:

- OCR accuracy
- Vulnerability matching
- LLM response time

## Limitations

- Depends on the completeness of the NVD database.
- Supports only a limited set of dependency formats.
- OCR accuracy depends on image quality.
- Complex version constraints may not always be fully matched.
- LLM-generated reports should be reviewed by a human before making security decisions.

## Fairness Considerations

The system evaluates software packages rather than people.

Potential biases may arise because:

- Some vendors publish more CVEs than others.
- Different software ecosystems have different vulnerability coverage.
- Older software generally has more reported vulnerabilities.

## Explainability

The main pipeline is retrieval-based and fully traceable.

Every reported vulnerability can be traced back to:

- the extracted package,
- the matched NVD record,
- the applied version constraint.

An auxiliary Random Forest classifier with SHAP analysis is included to demonstrate explainability for vulnerability-related features.

## Ethical Considerations

The system is intended to support security analysis rather than make autonomous decisions.

Generated reports should be treated as recommendations and validated by security professionals before deployment.
